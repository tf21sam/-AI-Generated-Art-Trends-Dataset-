```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Art_Style', y='Popularity_Score')
plt.title('Popularity Score by Art Style')
plt.xlabel('Art Style')
plt.ylabel('Popularity Score')
plt.xticks(rotation=90)
plt.show()
```

```python
from scipy.stats import ttest_ind
art_style1 = df[df['Art_Style'] == 'style1']['Popularity_Score']
art_style2 = df[df['Art_Style'] == 'style2']['Popularity_Score']
t_stat, p_value = ttest_ind(art_style1, art_style2)
print(f'T-statistic: {t_stat}, p-value: {p_value}')
```

```python
from sklearn.cluster import KMeans 
features = df[['Popularity_Score']]
KMeans = KMeans (n_clusters=3, random_state=0).fit(features)
df['Cluster'] = KMeans.labels_

plt.figure(figsize=(10, 6))
sns.scatterplot(data= df, x='Popularity_Score', y='Creation_Date', hue='Cluster', palette='viridis')
plt.title('K-means clustering of Artworks')
plt.xlabel('Popularity Score')
plt.ylabel('Creation Date')
plt.show()
```
